# 12.Software Development Vs Data Science(First Top Ten service Based Companies Jobs Openings) :-

```python
v_first_rows = v.head(10)
company_counts = v_first_rows.groupby('Companies')[['Data Science  ', 'Software Development ']].sum()

plt.figure(figsize=(12, 8))
bar_width = 0.35
index = np.arange(len(v_first_rows))

plt.bar(index, company_counts['Data Science  '], bar_width, label='Data Science')
plt.bar(index + bar_width, company_counts['Software Development '], bar_width, label='Software Development')

plt.xlabel('Company')
plt.ylabel('Count')
plt.title('Software Development Vs Data Science(First Top Ten service Based Companies Jobs Openings)')
plt.xticks(index + bar_width / 2, v_first_rows['Companies'], rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

# 16.Software Development Vs Data Science (Salary Comparison In Top Ten Service Based Companies):-

```python
data_science_salaries = v['Data Science(Salary)'][:10]
software_dev_salaries = v['Software Development (Salary)'][:10]
company_names = v['Companies'][:10]
x = np.arange(len(company_names))
fig, ax = plt.subplots(figsize=(12, 6))
bar_width = 0.35
bars1 = ax.bar(x - bar_width/2, data_science_salaries, bar_width, label='Data Science', color='black')
bars2 = ax.bar(x + bar_width/2, software_dev_salaries, bar_width, label='Software Development', color='white')
ax.set_xlabel('Company')
ax.set_ylabel('Salary')
ax.set_title('Salary Comparison for Data Science vs Software Development Positions')
ax.set_xticks(x)
ax.set_xticklabels(company_names, rotation=45, ha='right')
ax.legend()
ax.yaxis.set_label_text('Salary (INR)', fontsize=12)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x)))
plt.tight_layout()
plt.show()
```

# 18.Opening Salary Vs Highet Salary (In All commpanies):-

```python
openings = v['Opening Salary']
highest_salary = v['Highest Salary']
company_names = v['Companies']
x = np.arange(len(company_names))
fig, ax1 = plt.subplots(figsize=(12, 6))
bar_width = 0.35
bars1 = ax1.bar(x - bar_width/2, openings, bar_width, label='Openings', color='pink')
ax2 = ax1.twinx()
bars2 = ax2.bar(x + bar_width/2, highest_salary, bar_width, label='Highest Salary', color='yellow')
ax1.set_xlabel('Companies')
ax1.set_ylabel('Opening Salary (INR)')
ax2.set_ylabel('Highest Salary (INR)')
ax1.set_title('Comparison Of Opening Salary And Highest Salary In All Companies')
ax1.set_xticks(x)
ax1.set_xticklabels(company_names, rotation=60, ha='right')
ax1.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x)))
ax2.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x)))
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

# 20.Work From Home Facility Availability Comparison In All Companies (Pie Chart):-

```python
import matplotlib.pyplot as plt

# Assuming 'v' is your DataFrame containing interview difficulty data
work_from_home_counts = v['Work From Home '].value_counts()
colors = ['#ff9999', '#66b3ff']

plt.figure(figsize=(8, 6))
plt.pie(work_from_home_counts, labels=['{}: {}'.format(option, count) for option, count in zip(work_from_home_counts.index, work_from_home_counts)], autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Work From Home Facility Distribution')
plt.axis('equal')

plt.show()
```

# 21.Interview Difficulty Level Comparison In All Companies (Pie Chart):-
